```python

# TODO 05 - how much total sales, average sales, and standard deviation of sales your company make in 2017
#df[ df["Order Date"].dt.year == 2017 ].describe() this is code not include sum, so decided to use .agg()
df["Sales"][ df["Order Date"].dt.year == 2017 ].agg(["sum", "mean", "std"])
# sum sales = 484247.4981
# mean sales = 242.97415860511794
# std sales = 754.0533572593683
```

```python
# TODO 06 - which Segment has the highest profit in 2018
df[ df["Order Date"].dt.year == 2018 ].groupby("Segment")["Profit"].max()
#Consumer segment has the highest profit in 2018
```

```python
# TODO 07 - which top 5 States have the least total sales between 15 April 2019 - 31 December 2019
df[ (df["Order Date"] >= "2019-04-15") & (df["Order Date"] <= "2019-12-31") ]\
    .groupby("State")["Sales"]\
        .agg("sum")\
            .sort_values()

# New Hampshire < New Mexico < District of Columbia < Louisiana < South Carolina
```

```python
# TODO 08 - what is the proportion of total sales (%) in West + Central in 2019 e.g. 25%

# filterd West and Central data
filterd_west_central = df[ (df["Order Date"].dt.year == 2019) & (df["Region"].isin(["West", "Central"])) ]

# find the total sales for West and Central
total_sale_inWestCentral = filterd_west_central.groupby("Region")["Sales"].agg("sum").sum()

# find the total sales for all regions
total_sale_allRegion = df[ (df["Order Date"].dt.year == 2019) ].groupby("Region")["Sales"].agg("sum").sum()

# Calculate proportion of total sales in West and Central
Proportion_west_central = (total_sale_inWestCentral / total_sale_allRegion) * 100

print(f"Proportion of total sales in West and Central is {Proportion_west_central:.2f}%")
```

```python
# TODO 09 - find top 10 popular products in terms of number of orders vs. total sales during 2019-2020

# filtered data between 2019-2020
filtered_2019_2020 = df[ (df["Order Date"].dt.year >= 2019) & (df["Order Date"].dt.year <= 2020) ]

# Top 10 product of number of orders
top10_number_orders = filtered_2019_2020.groupby("Product Name").agg({"Order ID":"count"}).sort_values(by="Order ID", ascending=False).head(10).reset_index()

print(top10_number_orders)

# Top 10 product that total sales during 2019-2020
top10_total_sales = filtered_2019_2020.groupby("Product Name").agg({"Sales":"sum"}).sort_values(by="Sales", ascending=False).head(10).reset_index()

print(top10_total_sales)
```
